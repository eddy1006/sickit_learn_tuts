here we changed the value in newdf2 but still the change got reflected in newdf .why?
this happens because newdf2 is just a view of newdf that is it just points to newdf therefore any changes that take place in newdf2 will also appear in newdf

```python
newdf[0][1] = 'change'
```

```python
newdf.head()
```

```python
newdf2.head()
```

```python
newdf3 = newdf.copy()
```

```python
newdf3[0][2] = "not_visible"
```

```python
newdf3.head()
```

```python
newdf.head()
```

now here when we used the .copy() function with newdf3 it no longer will become a view of newdf instead it will copy all its content and will be an independent dataframe therefore changes made in newdf3 will not be visible in newdf

if we are setting a value in the dataframe then the above method of chaining which we were using (newdf[0][0]) is not recommended due to all that copy view stuff thats why we use loc function to do the same

```python
newdf.loc[0,0] = "BRUH"
```

```python
newdf.head(2)
```
